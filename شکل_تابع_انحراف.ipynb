```python
# فراخوانی کتابخانه های لازم
import pandas as pd
import math
import io
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# رسم تابع انحراف وانگ
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
landa=-2
for i in np.arange(0,1,0.01):
  A.append(i)
  B.append(norm.cdf(norm.ppf(i)- landa))

landa=-1
for i in np.arange(0,1,0.01):
  C.append(norm.cdf(norm.ppf(i)- landa))


landa=0
for i in np.arange(0,1,0.01):
  D.append(norm.cdf(norm.ppf(i)- landa))

landa=1
for i in np.arange(0,1,0.01):
  E.append(norm.cdf(norm.ppf(i)- landa))

landa=2
for i in np.arange(0,1,0.01):
  F.append(norm.cdf(norm.ppf(i)- landa))


fig, ax1= plt.subplots()
ax1.plot(A,B,'blue',label='landa=-2')
ax1.plot(A,C,'red',label='landa=-1')
ax1.plot(A,D,'pink',label='landa=0')
ax1.plot(A,E,'green',label='landa=1')
ax1.plot(A,F,'green',label='landa=2')
plt.legend()
plt.grid(True)
#plt.savefig("chart.png")
plt.show()
#files.download("chart.png")
```

```python
# رسم تابع انحراف توانی
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
c=0.1
for i in np.arange(0,1,0.01):
  A.append(i)
  B.append(i**(c))

c=0.3
for i in np.arange(0,1,0.01):
  C.append(i**(c))

c=0.5
for i in np.arange(0,1,0.01):
  D.append(i**(c))

c=0.7
for i in np.arange(0,1,0.01):
  E.append(i**(c))

c=0.9
for i in np.arange(0,1,0.01):
  F.append(i**(c))

fig, ax1= plt.subplots()
ax1.plot(A,B,'blue',label='c=0.1')
ax1.plot(A,C,'red',label='c=0.3')
ax1.plot(A,D,'pink',label='c=0.5')
ax1.plot(A,E,'green',label='c=0.7')
ax1.plot(A,F,'green',label='c=0.9')
plt.legend()
plt.grid(True)
plt.savefig("chart.png")
plt.show()
#files.download("chart.png")
```

```python
#  رسم تابع انحراف توانی دوگانه
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
b=1
for i in np.arange(0,1,0.01):
  A.append(i)
  B.append(1-(1-i)**b)

b=2
for i in np.arange(0,1,0.01):
  C.append(1-(1-i)**b)

b=3
for i in np.arange(0,1,0.01):
  D.append(1-(1-i)**b)

b=4
for i in np.arange(0,1,0.01):
  E.append(1-(1-i)**b)

b=5
for i in np.arange(0,1,0.01):
  F.append(1-(1-i)**b)

fig, ax1= plt.subplots()
ax1.plot(A,B,'blue',label='b=1')
ax1.plot(A,C,'red',label='b=2')
ax1.plot(A,D,'pink',label='b=3')
ax1.plot(A,E,'green',label='b=4')
ax1.plot(A,F,'green',label='b=5')
plt.legend()
plt.grid(True)
plt.savefig("chart.png")
plt.show()
#files.download("chart.png")
```
