```python
# تعریف رابطه 1-7-2
def p(y,l):
    t=1
    for i in range(0,l):
        t*=df2["p_x"][y+i]
    return t
```

```python
# تعریف رابطه 1-3-4
def p_star(y,l):
  return float(norm.cdf(norm.ppf(p(y,l))-0.1))
```

```python
# تعریف احتمال بقا دموآر

def p1(x,k,w):
  return (w-x-k)/(w-x)
```

```python
# تعریف احتمال بقا گومپرتز
def p2(x,k,B,C):
  return np.exp((-B/np.log(C))*(C**(x+k) - C**(x)))
```

```python
#تعریف احتمال بقا مکهم
def p3(x,k,A,B,C):
  return np.exp(-(A*k+(B/np.log(C))*(C**(x+k) - C**(x))))
```

```python
w=106
k=100
x=0
A=[]
B=[]
#demoivre
# w=106 و k=100 و x=0
for i in range(0,k+1):
  a=p1(x,i,106)
  A.append(x+i)
  B.append(a)
#Gompertz
C=[]
#B=0.0001
#C=1.1
for i in range(0,k+1):
  a=p2(x,i,0.0001,1.1)
  C.append(a)
#Meakham
#A=0.00257
#B=0.0000111955
#C=1.118
D=[]
for i in range(0,k+1):
  a=p3(x,i,0.00257,0.0000111955,1.118)
  D.append(a)
#Life Table
E=[]
for i in range(0,101):
  a=p(0,i)
  E.append(a)
#Wang_life_table
G=[]
for i in range(0,101):
   a=p_star(0,i)
   G.append(a)

fig, ax1= plt.subplots()
ax1.plot(A,B,'blue',label='de moirre')
ax1.plot(A,E,'red',label='life table')
#ax1.plot(A,G,'pink',label='Wang life table')
ax1.plot(A,C,'green',label='Gompertz')
ax1.plot(A,D,'gray',label='Meakham')
plt.legend()
plt.grid(True)
plt.savefig("chart.png")
plt.show()
#files.download("chart.png")
```
