```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto
from google.colab import files
```

```python
# پارامترهای توزیع پارتو
shape_param = 3 # پارامتر شکل (alpha)
scale_param = 800 # پارامتر مقیاس

# محدوده‌ی مقادیر x
x = np.linspace(scale_param,scale_param + 10 , 1000)

# محاسبه تابع چگالی احتمال (PDF) و تابع توزیع تجمعی (CDF)
pdf_values = pareto.pdf(x, shape_param, scale=scale_param)
cdf_values = pareto.cdf(x, shape_param, scale=scale_param)

# رسم تابع چگالی احتمال (PDF)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, 'b', label='PDF')
plt.title('Pareto Distribution - PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()

# رسم تابع توزیع تجمعی (CDF)
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, 'r', label='CDF')
plt.title('Pareto Distribution - CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()

# نمایش نمودار
plt.tight_layout()
#plt.savefig("chart.png")
plt.show()
#files.download("chart.png")
```

```python
# توزیع بتا
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# پارامترهای توزیع بتا
a, b = 2, 5 # پارامترهای شکل (alpha, beta)

# محدوده‌ی مقادیر x
x = np.linspace(0, 1, 1000)

# محاسبه تابع چگالی احتمال (PDF) و تابع توزیع تجمعی (CDF)
pdf_values = beta.pdf(x, a, b)
cdf_values = beta.cdf(x, a, b)

# رسم تابع چگالی احتمال (PDF)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, 'b', label='PDF')
plt.title('Beta Distribution - PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()

# رسم تابع توزیع تجمعی (CDF)
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, 'r', label='CDF')
plt.title('Beta Distribution - CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()

# نمایش نمودار
plt.tight_layout()
plt.savefig("chart.png")
plt.show()
files.download("chart.png")
```

```python
# لگ نرمال
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the log-normal distribution
mu = 0 # Mean of the log-normal distribution
sigma = 0.1 # Standard deviation of the log-normal distribution

# Range for plotting
x = np.linspace(0, 3, 1000)

# Calculate the probability density function (PDF)
pdf = stats.lognorm.pdf(x, s=sigma, scale=np.exp(mu))

# Calculate the cumulative distribution function (CDF)
cdf = stats.lognorm.cdf(x, s=sigma, scale=np.exp(mu))

# Plot PDF
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.title('Probability Density Function (PDF) of Log-Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid()
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='orange')
plt.title('Cumulative Distribution Function (CDF) of Log-Normal Distribution')
plt.xlabel('x')
plt.ylabel('Cumulative Distribution')
plt.grid()
plt.legend()

plt.tight_layout()
plt.savefig("chart.png")
plt.show()
files.download("chart.png")
```

```python
# نمایی
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the exponential distribution
lambda_param = 1 # Rate parameter (lambda)

# Range for plotting
x = np.linspace(0, 5, 1000)

# Calculate the probability density function (PDF)
pdf = stats.expon.pdf(x, scale=1/lambda_param)

# Calculate the cumulative distribution function (CDF)
cdf = stats.expon.cdf(x, scale=1/lambda_param)

# Plot PDF
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.title('Probability Density Function (PDF) of Exponential Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid()
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='orange')
plt.title('Cumulative Distribution Function (CDF) of Exponential Distribution')
plt.xlabel('x')
plt.ylabel('Cumulative Distribution')
plt.grid()
plt.legend()

plt.tight_layout()
plt.savefig("chart.png")
plt.show()
files.download("chart.png")
```

```python
# توزیع یکنواخت
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the continuous uniform distribution
a = 0 # Minimum value
b = 5 # Maximum value

# Range for plotting
x = np.linspace(a - 1, b + 1, 1000)

# Calculate the probability density function (PDF)
pdf = stats.uniform.pdf(x, loc=a, scale=(b - a))

# Calculate the cumulative distribution function (CDF)
cdf = stats.uniform.cdf(x, loc=a, scale=(b - a))

# Plot PDF
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.title('Probability Density Function (PDF) of Continuous Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.axvline(a, color='red', linestyle='--', label='a (min)')
plt.axvline(b, color='green', linestyle='--', label='b (max)')
plt.grid()
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='orange')
plt.title('Cumulative Distribution Function (CDF) of Continuous Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Cumulative Distribution')
plt.axvline(a, color='red', linestyle='--', label='a (min)')
plt.axvline(b, color='green', linestyle='--', label='b (max)')
plt.grid()
plt.legend()

plt.tight_layout()
plt.savefig("chart.png")
plt.show()
files.download("chart.png")
```
